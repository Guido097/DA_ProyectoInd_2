```python
hechos_h = pd.read_excel(df, 'HECHOS')
victimas_h = pd.read_excel(df,'VICTIMAS')
```

```python
#Creo una funcion para poder distinguir las columnas duplicadas.

def encontrar_columnas_duplicadas(df):
    columnas_unicas = []
    columnas_duplicadas = []

    for i, columna1 in enumerate(df.columns[:-1]):
        for columna2 in df.columns[i+1:]:
            if (df[columna1] == df[columna2]).all():
                columnas_duplicadas.append((columna1, columna2))
            else:
                columnas_unicas.append((columna1, columna2))

    return columnas_duplicadas, columnas_unicas

#Corro la funcion.
columnas_duplicadas, columnas_unicas = encontrar_columnas_duplicadas(df_combinado)

#Como solo nos interesan las columnas que se repiten, solo imprimimos estas
# de modo tal que sea facil eliminarlas.
print("Pares de columnas iguales:")
for columna1, columna2 in columnas_duplicadas:
    print(f"{columna1} y {columna2} son iguales")
```

```python
#Eliminamos las columnas que tienen valores duplicados luego del 'join'.

df_combinado = df_combinado.drop(['AAAA_y','MM_y'], axis=1)
```

```python
#Con lo visto anteriormente deberiamos tener las columnas 'DD', 'FECHA', 'VICTIMA' iguales,
# ya que son las columnas con el mismo nombre a las cuales se les agrega la 'x' o la 'y' para 
# identificarlas, pero no salen en la lista asi que vamos a checkear.

filas = []
for index, row in df_combinado[['FECHA_x','FECHA_y','DD_x','DD_y','VICTIMA_x','VICTIMA_y']].iterrows():
    if row['FECHA_x'] !=row['FECHA_y']:
        filas.append(index)
    if row['DD_x'] !=row['DD_y']:
        filas.append(index)
    if row['VICTIMA_x'] !=row['VICTIMA_y']:
        filas.append(index)

#Imprimimos las filas que no se repiten.

print(filas)
```

```python
#Imprimimos los valores de las columnas que checkeamos recien.
df_combinado[['FECHA_x','FECHA_y', 'DD_x','DD_y', 'VICTIMA_x',  'VICTIMA_y']].iloc[filas]
```

```python
#Como podemos ver las discrepancias en los valores de las columnas son minimas.


# En cuanto a las columnas "FECHA" y "DD", teniendo en cuenta que tenemos unificadas las 
# tablas de hechos y victimas la diferencia puede ser del hecho al dia de fallecimiento,
# y como para este trabajo nos interesan los hechos fatales para las mediciones a realizar,
# dejaremos las fechas de la tabla victimas que tiene los datos de los fallecimientos.

df_combinado = df_combinado.drop(['DD_x','FECHA_x'], axis=1)
```

```python
#Imprimimos las filas para corroborar los cambios.
df_combinado[['FECHA_y','DD_y', 'VICTIMA_x',  'VICTIMA_y', 'PARTICIPANTES','ACUSADO']].iloc[filas]
```

```python
#Vemos que solo queda una discrepancia queda en la fila 603 donde tenemos en la 
# columna 'VICTIMA_y' el valor 'MOTO' cuando deberia ser 'PEATON' ya que como vimos
# en la columna participantes los involucrados son un peaton y una moto, y no dos
# motos como figuraria en el dataframe, asi que realizamos ese cambio de la misma manera.

fila = 603
victima = 'PEATON'

df_combinado.at[fila, 'VICTIMA_y'] = victima
```

```python
#Corroboramos una vez mas antes de eliminar las columnas que no necesitamos.
df_combinado[['FECHA_y','DD_y', 'VICTIMA_x',  'VICTIMA_y', 'PARTICIPANTES','ACUSADO']].iloc[filas]
```

```python
#Habiendo corregido todo eliminamos la columna con los datos erroneos.

df_combinado = df_combinado.drop('VICTIMA_x', axis=1)
```

```python
#Creamos una funcion para reorganizar las direcciones de la columna "LUGAR DEL HECHO"

def reorg_direcc(address):
    words = address.split()
    if "Av." in words:
        words.remove("Av.")
        words = ["Av."] + words
    return " ".join(words)

df_combinado['LUGAR_DEL_HECHO'] = df_combinado['LUGAR_DEL_HECHO'].apply(reorg_direcc)
```

```python
#Seguimos corrigiendo los valores de la columna "LUGAR DEL HECHO"
df_combinado['LUGAR_DEL_HECHO'] =df_combinado['LUGAR_DEL_HECHO'].str.replace(',', '')
```

```python
#Borramos las que no nos hacen falta.

df_combinado = df_combinado.drop(['AAAA_x','MM_x','DD_y','Calle','XY (CABA)','Altura','Cruce','Dirección Normalizada'], axis=1)
```

```python
#Formateamos correctamente las columnas para normalizar las mayusculas.

df_combinado['ID'] = df_combinado['ID'].str.title()
df_combinado['ROL'] = df_combinado['ROL'].str.title()
df_combinado['VICTIMA_y'] = df_combinado['VICTIMA_y'].str.title()
df_combinado['SEXO'] = df_combinado['SEXO'].str.title()
df_combinado['LUGAR_DEL_HECHO'] = df_combinado['LUGAR_DEL_HECHO'].str.title()
df_combinado['TIPO_DE_CALLE'] = df_combinado['TIPO_DE_CALLE'].str.title()
df_combinado['pos x'] = df_combinado['pos x'].str.title()
df_combinado['pos y'] = df_combinado['pos y'].str.title()
df_combinado['PARTICIPANTES'] = df_combinado['PARTICIPANTES'].str.title()
df_combinado['ACUSADO'] = df_combinado['ACUSADO'].str.title()
df_combinado['ROL'] = df_combinado['ROL'].str.title()
```

```python
 # Renombramos algunas columnas que quedaron con el nombre cambiado por el JOIN.


df_combinado = df_combinado.rename(columns={'VICTIMA_y': 'VICTIMA'})
df_combinado = df_combinado.rename(columns={'FECHA_y': 'FECHA'})
```

```python
# Quitamos del dataframe a las filas con faltantes en las columnas principales.

df_combinado = df_combinado[(df_combinado['PARTICIPANTES'] != 'Sd') & (df_combinado['VICTIMA'] != 'Sd') & (df_combinado['ACUSADO'] != 'Sd')]
```

```python
# Hacemos un conteo de los valores "SD" del dataframe.

conteo_sd = df_combinado.apply(lambda x: x.value_counts().get('SD', 0))

print(conteo_sd)
```

```python
df_combinado
```

```python
#Como hay faltantes en la columna de "FECHA FALLECIMIENTO", se va a rellenar con la columna "FECHA".
# Para esto, pasamos el formato a 'str' para facilitar la tarea.

df_combinado['FECHA_FALLECIMIENTO'] = df_combinado['FECHA_FALLECIMIENTO'].astype(str)
df_combinado['FECHA'] = df_combinado['FECHA'].astype(str)
df_combinado.loc[df_combinado['FECHA_FALLECIMIENTO'] == 'SD', 'FECHA_FALLECIMIENTO'] = df_combinado['FECHA']
df_combinado['FECHA_FALLECIMIENTO'] = df_combinado['FECHA_FALLECIMIENTO'].str.replace('T', ' ')
```

```python
#Para evitar malentendidos renombramos algunas columnas.

df_combinado = df_combinado.rename(columns={'HORA': 'HORA_ACCIDENTE'})
df_combinado = df_combinado.rename(columns={'FECHA': 'FECHA_ACCIDENTE'})
df_combinado = df_combinado.rename(columns={'pos x': 'LONGITUD'})
df_combinado = df_combinado.rename(columns={'pos y': 'LATITUD'})
```

```python
# Extraer las horas y minutos de las cadenas de texto
df_combinado['HORA'] = df_combinado['HORA_ACCIDENTE'].str.split(':').str[0]
df_combinado['MINUTO'] = df_combinado['HORA_ACCIDENTE'].str.split(':').str[1]

# Se convierten las horas y minutos a números
df_combinado['HORA'] = pd.to_numeric(df_combinado['HORA'], errors='coerce')
df_combinado['MINUTO'] = pd.to_numeric(df_combinado['MINUTO'], errors='coerce')

# Calculamos la media de las horas y minutos.
mean_hora = df_combinado['HORA'].mean()
mean_minuto = df_combinado['MINUTO'].mean()

# Crear la media de la hora como una cadena de texto
mean_time = f"{int(mean_hora)}:{int(mean_minuto)}"

# Reemplazamos los valores "SD" y "Sd" con la media de la hora, que creamos antes.
df_combinado.loc[df_combinado['HORA_ACCIDENTE'] == 'SD', 'HORA_ACCIDENTE'] = mean_time
df_combinado.loc[df_combinado['HORA_ACCIDENTE'] == 'Sd', 'HORA_ACCIDENTE'] = mean_time

#Ahora se eliminan las columnas 'HORA' y 'MINUTO'
df_combinado = df_combinado.drop(columns=['HORA', 'MINUTO'])
```

```python
df_combinado
```

```python
#Le damos formato fecha a la columna "FECHA_ACCIDENTE" para poder sacar el semestre en el que ocurrio.
df_combinado['FECHA_ACCIDENTE'] = pd.to_datetime(df_combinado['FECHA_ACCIDENTE'])
```

```python
# Ahora creamos la columna "SEMESTRE" y de acuerdo al mes de "FECHA_ACCIDENTE" completamos los valores.
df_combinado['SEMESTRE'] = [1 if month <= 6 else 2 for month in df_combinado['FECHA_ACCIDENTE'].dt.month]
```

```python
# Creamos la columna "ANIO" para facilitar el calculo del KPI

df_combinado['ANIO'] = df_combinado['FECHA_ACCIDENTE'].dt.year
```

```python
#Eliminamos la columna "HH" ya que no aporta nada.
df_combinado = df_combinado.drop(['HH'], axis=1)
```

```python
#Formateamos en formato numerico la columna "EDAD". Dejando los valores "SD" como nulos.

df_combinado['EDAD'] = pd.to_numeric(df_combinado['EDAD'], errors='coerce')

# Calculamos la media de la columna edad para poder completar los datos faltantes de la misma.

mean_edad = df_combinado['EDAD'].mean()
mean_edad = int(mean_edad)

#Ahora rellenamos los nulos con la media de la edad.

df_combinado['EDAD'] = df_combinado['EDAD'].fillna(mean_edad)

#Ahora formateamos los valores de la columna para que queden numeros enteros.

df_combinado['EDAD'] = df_combinado['EDAD'].astype(int)
```
