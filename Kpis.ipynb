```python
import pandas as pd
import numpy as np
```

Calculo KPI nro1. 

-Reducir en un 10% la tasa de homicidios en siniestros viales de los últimos seis meses, en CABA, en comparación con la tasa de homicidios en siniestros viales del semestre anterior.

Definimos a la tasa de homicidios en siniestros viales como el número de víctimas fatales en accidentes de tránsito por cada 100,000 habitantes en un área geográfica durante un período de tiempo específico. Su fórmula es: (Número de homicidios en siniestros viales / Población total) * 100,000

```python
#Cargamos el '.csv' con los datos de los accidentes que trabajamos antes.
homicidios = pd.read_csv('DataBase/homicidios_final.csv')
#Cargamos el '.csv' con la poblacion aproximada de acuerdo a los censos.
poblacion = pd.read_csv('DataBase/poblacionCABA.csv')
```

```python
homicidios
```

```python
#Aislamos los datos de victimas totales por semestre y anio.
homicidios_sum = homicidios.groupby(['ANIO', 'SEMESTRE'])['N_VICTIMAS'].sum().reset_index()
```

```python
homicidios_sum
```

```python
poblacion
```

```python
#Eliminamos las columnas que nos son innecesarias.
poblacion = poblacion.drop(columns=['Unnamed: 0'])
poblacion = poblacion.drop(columns=['Poblacion 2010'])
poblacion = poblacion.drop(columns=['Crecimiento 2010-2022'])
```

```python
poblacion_sum = poblacion[['Poblacion 2016', 'Poblacion 2017', 'Poblacion 2018', 'Poblacion 2019', 'Poblacion 2020', 'Poblacion 2021']].sum()
```

```python
poblacion_sum_df = poblacion_sum.to_frame().reset_index()
poblacion_sum_df.columns = ['ANIO', 'POBLACION TOTAL']
```

```python
poblacion_sum_df
```

```python
poblacion_sum_df['ANIO'] = poblacion_sum_df['ANIO'].str.replace('Poblacion', '', case=False)
```

```python
poblacion_sum_df
```

```python
#Formateamos en enteros la columna "ANIO" de ambos dataframe para facilitar el JOIN.

poblacion_sum_df['ANIO'] = poblacion_sum_df['ANIO'].astype(int)
homicidios_sum['ANIO'] = homicidios_sum['ANIO'].astype(int)
```

```python
#Fusionamos los dos dataframes que estabamos trabajando.

kpi_1 = homicidios_sum.merge(poblacion_sum_df, on='ANIO')
```

```python
kpi_1
```

```python
#Ahora pasamos a calcular el kpi.

kpi_1['TASA'] = (kpi_1['N_VICTIMAS'] / kpi_1['POBLACION TOTAL']) * 100000
```

```python
kpi_1
```

```python
#Creamos la columna "TASA_ANT" y le damos el valor de la columna "TASA" pero
# desplazado una fila mas abajo.
kpi_1['TASA_ANT'] = kpi_1['TASA'].shift(periods=1, fill_value=0)
kpi_1
```

```python
#Calculamos el porcentaje de variacion de la tasa con respecto al semestre anterior.
kpi_1['VAR_TASA'] = (kpi_1['TASA'] - kpi_1['TASA_ANT']) / kpi_1['TASA_ANT'] * 100
kpi_1
```

```python
#Reemplazamos el valor 'inf' por '0'.
kpi_1['VAR_TASA'] = kpi_1['VAR_TASA'].replace([np.inf, -np.inf], 0)
kpi_1
```

```python
#Redondeamos el porcentaje de variacion de la tasa de homicidios.
kpi_1['VAR_TASA'] = kpi_1['VAR_TASA'].round(2)
kpi_1
```

Calculo KPI Nro2.

-Reducir en un 7% la cantidad de accidentes mortales de motociclistas en el último año, en CABA, respecto al año anterior.

Definimos a la cantidad de accidentes mortales de motociclistas en siniestros viales como el número absoluto de accidentes fatales en los que estuvieron involucradas víctimas que viajaban en moto en un determinado periodo temporal. Su fórmula para medir la evolución de los accidentes mortales con víctimas en moto es: (Número de accidentes mortales con víctimas en moto en el año anterior - Número de accidentes mortales con víctimas en moto en el año actual) / (Número de accidentes mortales con víctimas en moto en el año anterior) * 100

```python
homicidios_moto = homicidios[homicidios['VICTIMA'] == 'Moto']
homicidios_moto
```

```python
#Aislamos los datos de victimas fatales, donde la victima es una moto, por anio.
homicidios_moto_sum = homicidios_moto.groupby(['ANIO'])['N_VICTIMAS'].sum().reset_index()
homicidios_moto_sum
```

```python
#Como hicimos en el kpi anterior, creamos una columna "N_VICTIMAS_ANT" donde ubicamos
# el dato de "N_VICTIMAS" desplazados una fila.
homicidios_moto_sum['N_VICTIMAS_ANT'] = homicidios_moto_sum['N_VICTIMAS'].shift(periods=1, fill_value=0)
homicidios_moto_sum
```

```python
#Calculamos el porcentaje de variacion anual de victimas fatales.
homicidios_moto_sum['VARIACION'] = (homicidios_moto_sum['N_VICTIMAS'] - homicidios_moto_sum['N_VICTIMAS_ANT']) / homicidios_moto_sum['N_VICTIMAS_ANT'] * 100
homicidios_moto_sum
```

```python
kpi_2 = homicidios_moto_sum
```

```python
#Reemplazamos el valor 'inf' por '0'.

kpi_2['VARIACION'] = kpi_2['VARIACION'].replace([np.inf, -np.inf], 0)
kpi_2
```

```python
#Redondeamos el porcentaje de variacion del numero de victimas fatales.
kpi_2['VARIACION'] = kpi_2['VARIACION'].round(2)
kpi_2
```
